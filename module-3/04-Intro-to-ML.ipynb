##ㅁlgunos conceptos
Para entender c칩mo un algoritmo de aprendizaje autom치tico aprende de los datos para predecir un resultado, es esencial comprender los conceptos subyacentes que intervienen en el entrenamiento de un algoritmo.

### Estimadores
La estimaci칩n es un t칠rmino estad칤stico para encontrar alguna estimaci칩n de un par치metro desconocido, dados algunos datos. La estimaci칩n puntual es el intento de proporcionar la mejor predicci칩n de alguna cantidad de inter칠s.
La cantidad de inter칠s puede ser: 
 - Un 칰nico par치metro
 - Un vector de par치metros - por ejemplo, los pesos en la regresi칩n lineal 
 - Una funci칩n completa

###Error de Bias

Es la diferencia entre la predicci칩n esperada de nuestro modelo y los valores verdaderos. Aunque al final nuestro objetivo es siempre construir modelos que puedan predecir datos muy cercanos a los valores verdaderos, no siempre es tan f치cil porque algunos algoritmos son simplemente demasiado r칤gidos para aprender se침ales complejas del conjunto de datos.

Imagina ajustar una regresi칩n lineal a un conjunto de datos que tiene un patr칩n no lineal, no importa cu치ntas observaciones m치s recopiles, una regresi칩n lineal no podr치 modelar las curvas en esos datos. Esto se conoce como ajuste insuficiente.

En general, los algoritmos param칠tricos como la regresi칩n lineal, tienen un alto bias que los hace r치pidos de aprender y m치s f치cil de entender, pero generalmente menos flexibles. A su vez, tienen un menor rendimiento predictivo en problemas complejos.

###Error de Varianza

Se refiere a la cantidad que la estimaci칩n de la funci칩n objetivo cambiar치 si se utiliza diferentes datos de entrenamiento. La funci칩n objetivo se estima a partir de los datos de entrenamiento mediante un algoritmo de Machine Learning, por lo que deber칤amos esperar que el algoritmo tenga alguna variaci칩n. Idealmente no deber칤a cambiar demasiado de un conjunto de datos de entrenamiento a otro, lo que significa que el algoritmo es bueno para elegir el mapeo subyacente oculto entre las variables de entrada y de salida.

Los algoritmos de Machine Learning que tienen una gran varianza est치n fuertemente influenciados por los detalles de los datos de entrenamiento, esto significa que los detalles de la capacitaci칩n influyen en el n칰mero y los tipos de par치metros utilizados para caracterizar la funci칩n de mapeo.

Generalmente, los algoritmos de Machine Learning no param칠trico que tienen mucha flexibilidad tienen una gran variaci칩n.

- Varianza baja: sugiere peque침os cambios en la estimaci칩n de la funci칩n objetivo con cambios en el conjunto de datos de capacitaci칩n.
- Alta varianza: sugiere grandes cambios en la estimaci칩n de la funci칩n objetivo con cambios en el conjunto de datos de capacitaci칩n.
Los algoritmos de Machine Learning con baja varianza incluyen: regresi칩n lineal, an치lisis discriminante lineal y regresi칩n log칤stica.

Por su parte, los algoritmos con alta varianza son: 치rboles de decisi칩n, k-vecinos m치s cercanos y m치quinas de vectores de soporte.

## Tipos de Aprendizaje Autom치tico
**- Supervisado**      
El modelo se entrena con datos etiquetados, es decir, con datos cuya verdad b치sica se conoce. Cuando se le presentan datos y una etiqueta, el modelo infiere patrones.      
**- No supervisado**          
No hay verdades b치sicas. El modelo busca patrones no detectados previamente, con los que separar los distintos puntos de datos en diferentes clusters.
**- Refuerzo**    
Tampoco existe una verdad de base. La acci칩n del modelo se valora y se da una recompensa o un castigo en consecuencia. El objetivo del modelo es obtener el mayor n칰mero de recompensas posible.

## Features
Las **Features** son variables individuales independientes que act칰an como entrada en su sistema. Los modelos de predicci칩n utilizan caracter칤sticas para
hacer predicciones.      
**Objetivo**        
El objetivo es la salida de las variables de entrada. Pueden ser las clases individuales a las que se asignan las variables de entrada en el caso de un problema de clasificaci칩n o el rango de valores de salida en un problema de regresi칩n.

## Underfitting y Overfitting  
Dividir un conjunto de datos tambi칠n puede ser importante para detectar si su modelo sufre uno de los dos problemas m치s comunes, denominados infraajuste y sobreajuste:

El **underfitting** suele ser la consecuencia de que un modelo sea incapaz de encapsular las relaciones entre los datos. Por ejemplo, esto puede ocurrir cuando se intenta representar relaciones no lineales con un modelo lineal. Los modelos infraajustados probablemente tendr치n un rendimiento pobre tanto en los conjuntos de entrenamiento como en los de prueba.

El **overfitting** suele tener lugar cuando un modelo tiene una estructura excesivamente compleja y aprende tanto las relaciones existentes entre los datos como el ruido. Estos modelos suelen tener una mala capacidad de generalizaci칩n. Aunque funcionan bien con los datos de entrenamiento, suelen tener un rendimiento pobre con los datos no vistos (de prueba).

###Train / Test Split 쮼sto qu칠 es?

![eldibujo](images/traintest.png)

La divisi칩n entrenamiento-prueba es una t칠cnica para evaluar el rendimiento de un algoritmo de aprendizaje autom치tico.

Puede utilizarse para problemas de clasificaci칩n o regresi칩n y puede emplearse para cualquier algoritmo de aprendizaje supervisado.

El procedimiento consiste en tomar un conjunto de datos y dividirlo en dos subconjuntos. El primer subconjunto se utiliza para ajustar el modelo y se denomina conjunto de datos de entrenamiento. El segundo subconjunto no se utiliza para entrenar el modelo; en su lugar, se proporciona al modelo el elemento de entrada del conjunto de datos, luego se hacen predicciones y se comparan con los valores esperados. Este segundo conjunto de datos se denomina conjunto de datos de prueba.

Conjunto de datos de entrenamiento: Se utiliza para ajustar el modelo de aprendizaje autom치tico.
Conjunto de datos de prueba: Se utiliza para evaluar el modelo de aprendizaje autom치tico ajustado.
El objetivo es estimar el rendimiento del modelo de aprendizaje autom치tico con datos nuevos: datos no utilizados para entrenar el modelo.

As칤 es como esperamos utilizar el modelo en la pr치ctica. Es decir, ajustarlo sobre los datos disponibles con entradas y salidas conocidas, para luego hacer predicciones sobre nuevos ejemplos en el futuro en los que no tenemos los valores de salida u objetivo esperados.

El procedimiento de entrenamiento-prueba es adecuado cuando se dispone de un conjunto de datos suficientemente amplio.

### Preparaci칩n de los datos para train test split

### Par치metros opcionales del train test split

**train_size** es el n칰mero que define el tama침o del conjunto de entrenamiento. Si proporciona un float, entonces debe estar entre 0.0 y 1.0 y definir치 la parte del conjunto de datos utilizado para las pruebas. Si proporciona un int, entonces representar치 el n칰mero total de las muestras de entrenamiento. El valor por defecto es None.

**test_size** es el n칰mero que define el tama침o del conjunto de pruebas. Es muy similar a train_size. Debe proporcionar o bien train_size o bien test_size. Si no se proporciona ninguno de los dos, la parte por defecto del conjunto de datos que se utilizar치 para las pruebas es 0,25, o el 25 por ciento.

**random_state** es el objeto que controla la aleatoriedad durante la divisi칩n. Puede ser un int o una instancia de RandomState. El valor por defecto es None.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

### Ya tenemos el dataset dividido, ahora vamos a importar un modelo y entrenarlo

```python
from sklearn.linear_model import LinearRegression
```

```python
#Creamos el objeto LinearRegression en una variable
lr = LinearRegression()
```

```python
# Entrenamos el modelo
lr.fit(X_train, y_train)
```

###Creamos nuestras predicciones

```python
y_pred = lr.predict(X_test)
```

```python
prueba = X[0:1]
```

### Medimos nuestro modelo

```python
from sklearn import metrics
import numpy as np
```

```python
print("MAE: ", metrics.mean_absolute_error(y_test,y_pred))
print("MSE: ", metrics.mean_squared_error(y_test,y_pred))
print("RMSE: ", np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print("R2: ", metrics.r2_score(y_test, y_pred))
```

** C치lculo del error medio absoluto, del error medio cuadr치tico y del error medio cuadr치tico

- MAE es el m치s f치cil de entender, porque es el error medio.
- El MSE es m치s popular que el MAE, porque el MSE "castiga" los errores m치s grandes, lo que suele ser 칰til en el mundo real.
- RMSE es a칰n m치s popular que MSE, es la ra칤z cuadrada del MSE y mide la desviaci칩n est치ndar de los residuos.
Todas estas son funciones de p칠rdida, porque queremos minimizarlas.

## Extra! Entrenando diferentes modelos a la vez

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.linear_model import SGDRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

### Almacenamos todos los modelos que vamos a entrenar en un diccionario

```python
models = { "ridge": Ridge(),
          "lasso": Lasso(),
          "sgd": SGDRegressor(),
          "knn": KNeighborsRegressor(),
          "gradient": GradientBoostingRegressor()
}
```

### Iteramos sobre los modelos para entrenarlos

```python
for name, model in models.items():
    print(f"Estoy entrenando el modelo ---> {name}")
    model.fit(X_train, y_train)
    print("He acabado")
```

```python
#Podemos seguir el mismo proceso para realizar predicciones de cada algoritmo y sacar m칠tricas
for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"--------{name}--------")
    print("MAE: ", metrics.mean_absolute_error(y_test,y_pred))
    print("MSE: ", metrics.mean_squared_error(y_test,y_pred))
    print("RMSE: ", np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
    print("R2: ", metrics.r2_score(y_test, y_pred))
```
